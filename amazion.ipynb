```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("amazon.csv")
df
```

```python
if df.isnull().sum().sum() > 0:
    df.fillna(df.median(numeric_only=True), inplace=True)  # Fill numeric columns with median
    df.fillna(df.mode().iloc[0], inplace=True)  # Fill categorical columns with mode

if df.duplicated().sum() > 0:
    df.drop_duplicates(inplace=True)
```

```python
#Q1. What is the average rating for each product category?
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

category_avg_rating = df.groupby("category")["rating"].mean().sort_values(ascending=False)
print(category_avg_rating)
```

```python
#Q2.What are the top rating_count products by category?
category_avg_rating = df.groupby("category")["rating"].mean().sort_values(ascending=False)
print(category_avg_rating)
```

```python
#Q3.What is the distribution of discounted prices vs. actual prices?
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")

df = df.dropna(subset=["rating_count"])

df["rating_count"] = df["rating_count"].astype(int)

top_rated_products = df.groupby("category").apply(lambda x: x.nlargest(5, "rating_count"))[["product_name", "rating_count"]]
print(top_rated_products)
```

```python
#Q4. How does the average discount percentage vary across categories?
plt.figure(figsize=(10, 5))
sns.histplot(df["actual_price"], bins=30, color="red", label="Actual Price", alpha=0.5)
sns.histplot(df["discounted_price"], bins=30, color="blue", label="Discounted Price", alpha=0.5)
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Distribution of Discounted vs. Actual Prices")
plt.legend()
plt.show()
```

```python
#Q5. What are the most popular product names?
df["actual_price"] = pd.to_numeric(df["actual_price"], errors="coerce")
df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors="coerce")

df["discount_percentage"] = ((df["actual_price"] - df["discounted_price"]) / df["actual_price"]) * 100

category_avg_discount = df.groupby("category")["discount_percentage"].mean().sort_values(ascending=False)
print(category_avg_discount)
```

```python
#Q6. What are the most popular product keywords?
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")

popular_products = df.nlargest(10, "rating_count")[["product_name", "rating_count"]]
print(popular_products)
```

```python
#Q7. What are the most popular product reviews?
from collections import Counter
import itertools

# Tokenizing product names
keywords = list(itertools.chain(*[name.split() for name in df["product_name"].dropna()]))
common_keywords = Counter(keywords).most_common(10)

print("Top 10 Keywords:", common_keywords)
```

```python
#Q8. 
popular_reviews = df.nlargest(10, "rating")[["product_name", "rating", "review_title"]]
print(popular_reviews)
```

```python
#Q9.
plt.figure(figsize=(14, 8))
sns.scatterplot(x=df["discounted_price"], y=df["rating"])
plt.xlabel("Discounted Price")
plt.ylabel("Rating")
plt.title("Correlation Between Discounted Price and Rating")
plt.show()
```

```python
#Q10. 
top_categories = category_avg_rating.head(5)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_categories.index, y=top_categories.values, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.title("Top 5 Categories Based on Highest Ratings")
plt.show()
```
