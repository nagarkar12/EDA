```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Car Sale.csv")
df
```

```python
#Q1.  What is the average selling price of cars for each dealer, and how does it compare across different dealersE
avg_price = df.groupby('Dealer_Name')['Price ($)'].mean()
print(avg_price)
```

```python
#Q2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends
price_variation = df.groupby('Company')['Price ($)'].std()

# Plot the price variation across brands
price_variation.sort_values(ascending=False).plot(kind='bar', figsize=(12, 6), color='skyblue', edgecolor='black')
plt.xlabel("Car Brand")
plt.ylabel("Price Variation (Standard Deviation)")
plt.title("Price Variation Across Car Brands")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--")
plt.show()
```

```python
#Q3. What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare
price = df.groupby('Transmission')['Price ($)'].describe()
print(price)
```

```python
#Q4. What is the distribution of car prices across different regions?
regions = df.groupby('Dealer_Region')['Price ($)'].describe()
print(regions)
```

```python
#Q5. What is the distribution of cars based on body stylesE
styles = df['Body Style'].value_counts()
print(styles)
```

```python
#Q6. How does the average selling price of cars vary by customer gender and annual incomeE
gender_avg = df.groupby('Gender')['Price ($)'].mean()

gender_avg.plot(kind='bar', color=['blue', 'pink'], figsize=(8, 5))
plt.title("Average Car Selling Price by Gender")
plt.ylabel("Average Price ($)")
plt.xlabel("Gender")
plt.grid(axis="y", linestyle="--")
plt.show()
```

```python
#Q7. What is the distribution of car prices by region, and how does the number of cars sold vary by regionE
sale = df.groupby('Dealer_Region')['Price ($)'].agg(['count', 'mean'])
print(sale)
```

```python
#Q8. How does the average car price differ between cars with different engine sizes?
avg = df.groupby('Engine')['Price ($)'].mean()
print(avg)
```

```python
#Q9. How do car prices vary based on the customer’s annual income bracketE
avg_price = df.groupby('Annual Income')['Price ($)'].mean()

plt.figure(figsize=(12, 5))
avg_price.plot(marker='o', linestyle='-', color='green')
plt.xlabel("Annual Income Bracket")
plt.ylabel("Average Car Price ($)")
plt.title("Average Car Price Across Annual Income Brackets")
plt.grid()
plt.show()
```

```python
#Q10. What are the top 5 car models with the highest number of sales, and how does their price distribution look?
top = df['Model'].value_counts().head(5)

top_price = df[df['Model'].isin(top.index)].groupby('Model')['Price ($)'].describe()

plt.figure(figsize=(8, 5))
top.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel("Car Model")
plt.ylabel("Number of Sales")
plt.title("Top 5 Best-Selling Car Models")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--")
plt.show()
```

```python
#Q11. How does car price vary with engine size across different car colors, and which colors have the highest price variation
pivot_table = df.pivot_table(values='Price ($)', index='Engine', columns='Color', aggfunc='std')

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="coolwarm", linewidths=0.5)
plt.title("Price Variation Across Colors & Engine Sizes")
plt.ylabel("Engine Size")
plt.xlabel("Car Color")
plt.show()
```

```python
#Q12. Is there any seasonal trend in car sales based on the date of sale?
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

month = df.groupby('Month')['Model'].count()
print(month)
```

```python
#Q13. How does the car price distribution change when considering different combinations of body style and transmission type
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['Body Style'], y=df['Price ($)'], hue=df['Transmission'], palette="coolwarm")
plt.xlabel("Body Style")
plt.ylabel("Price ($)")
plt.title("Car Price Distribution by Body Style and Transmission")
plt.xticks(rotation=45)
plt.legend(title="Transmission")
plt.grid(axis="y", linestyle="--")
plt.show()
```

```python
#Q14. What is the correlation between car price, engine size, and annual income of customers, and how do these features interact
sns.pairplot(df[['Price ($)', 'Engine', 'Annual Income']])
plt.show()
```

```python
#Q15. How does the average car price vary across different car models and engine types?
engine_avg_price = df.groupby('Engine')['Price ($)'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
engine_avg_price.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel("Engine Type")
plt.ylabel("Average Price ($)")
plt.title("Average Car Price Across Engine Types")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--")
plt.show()
```
